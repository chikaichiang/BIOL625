```python
import gzip

input_filename1 = 'uniprotkb_proteome_UP000001926_2025_05_31.fasta.gz'
output_filename1 = 'entamoeba_histolytica_protoeme.faa'

# Decompress and save
with gzip.open(input_filename1, 'rt') as f_in, open(output_filename1, 'w') as f_out:
    for line in f_in:
        f_out.write(line)

print(f'File saved as: {output_filename1}')
files.download(output_filename1)
```

```python
count = 0
with open(output_filename1, 'r') as f:
    for line in f:
        if line.startswith('>'):
            count += 1

print(f'Total sequences in {output_filename1}: {count}')
```
